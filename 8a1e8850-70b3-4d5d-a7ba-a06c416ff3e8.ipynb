```python
from scipy import stats as st
alpha = 0.05

xb1 = df[(df['platform'] == 'XOne') & (df['user_score'] > 0) & (df['year_of_release'] >= 2010)]['user_score']
pc = df[(df['platform'] == 'PC') & (df['user_score'] > 0) & (df['year_of_release'] >= 2010)]['user_score']

results = st.ttest_ind(xb1, pc)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Для проверок гипотез используем t-критерий Стьюдента. Имеем 2 гипотезы:

Н0:Средние пользовательские рейтинги жанров Action и Sports схожи

Н1:Средние пользовательские рейтинги жанров Action и Sports разнятся

```python
act = df[(df['genre'] == 'Action') & (df['user_score'] > 0)& (df['year_of_release'] >= 2010)]['user_score']
sprt = df[(df['genre'] == 'Sports') & (df['user_score'] > 0)& (df['year_of_release'] >= 2010)]['user_score']

results = st.ttest_ind(act, sprt)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

<div class="alert alert-danger">
<font size="5"><b>Комментарий ревьюера</b></font>

Нулевая гипотеза всегда формулируется с учетом знака равенства. При получении pvalue меньше alpha, мы отклоняем нулевую гипотезу. Исправь данные неточности.
